```python
from unsloth import FastLanguageModel

MODEL_PATH = "/mnt/raid6/jaewook133/models/Llama-3.2-1B-Inst_16bit_translate_en-ko"
# MODEL_PATH = "unsloth/DeepSeek-R1-Distill-Qwen-1.5B"

hf_path = "Jooinjang/{}".format(MODEL_PATH.split("models/")[-1]).replace("_16bit", "")
```

```python
hf_path
```

```python
model, tokenizer = FastLanguageModel.from_pretrained(
        model_name=MODEL_PATH, max_length=4096, load_in_4bit=True, dtype=None
    )
```

```python
tokenizer.chat_template
```

```python
model.push_to_hub_gguf(hf_path, 
                       tokenizer=tokenizer,
                       quantization_method="q4_k_m",
)
print("q4_k_m done")
```

```python
model.push_to_hub_gguf(hf_path, 
                       tokenizer=tokenizer,
                       quantization_method="q8_0",
)
print("q8_0 done")
```

```python
model.push_to_hub_gguf(hf_path, 
                       tokenizer=tokenizer,
                       quantization_method="f16",
)
print("f16 done")
```
